Observation: Exploratory Analysis (Products Dataframe)

An exploratory review of the df_products dataframe revealed outliers in the prices column.

The column showed a mean of 9.99 with a standard deviation of 453.6, which is far beyond two standard deviations from the mean—clear evidence of outliers.

Specifically, two extreme values were identified: 14,900 and 99,999.

To correct this, both were replaced with the median value of 7.1, since the median is less influenced by extreme values than the mean.

This adjustment reduced the column’s mean to 7.6.

Considering the wide spread of the data (min = 1, max = 99,999), using the median ensures a more stable and representative measure for price.

Therefore, it is not necessary to remove these missing values from the dataset.

Observation 6. Missing Values
